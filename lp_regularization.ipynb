```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
init_x = np.random.randint(-100, 100, size=10)
init_y = np.random.randint(-100, 100, size=10)
step_size = 2
eps = 1e-9
```

```python
plt.figure(figsize=(16, 16))

# L1
plt.subplot(221)
plt.axhline(y=0, color="black", linestyle="-")
plt.axvline(x=0, color="black", linestyle="-")
plt.xlim(-100, 100)
plt.ylim(-100, 100)

for i in range(len(init_x)):
    x, y = [init_x[i]], [init_y[i]]

    for _ in range(100):
        dx = x[-1] / (abs(x[-1]) + eps) 
        dy = y[-1] / (abs(y[-1]) + eps)
        x.append(x[-1] - dx * step_size)
        y.append(y[-1] - dy * step_size)
        if np.sqrt(x[-1]**2 + y[-1]**2) < 1:
            break

    plt.plot(x, y, '.', markersize=5)

plt.xlabel("L1 regularization", fontsize=14)

# L2
plt.subplot(222)
plt.axhline(y=0, color="black", linestyle="-")
plt.axvline(x=0, color="black", linestyle="-")
plt.xlim(-100, 100)
plt.ylim(-100, 100)

for i in range(len(init_x)):
    x, y = [init_x[i]], [init_y[i]]

    for _ in range(100):
        dx = x[-1] / (np.sqrt(x[-1] ** 2 + y[-1] ** 2) + eps)
        dy = y[-1] / (np.sqrt(x[-1] ** 2 + y[-1] ** 2) + eps)
        x.append(x[-1] - dx * step_size)
        y.append(y[-1] - dy * step_size)
        if np.sqrt(x[-1]**2 + y[-1]**2) < 1:
            break
            
    plt.plot(x, y, '.', markersize=5)

plt.xlabel("L2 regularization", fontsize=14)

# L3
plt.subplot(223)
plt.axhline(y=0, color="black", linestyle="-")
plt.axvline(x=0, color="black", linestyle="-")
plt.xlim(-100, 100)
plt.ylim(-100, 100)

for i in range(len(init_x)):
    x, y = [init_x[i]], [init_y[i]]
    for _ in range(100):
        dx = (x[-1] ** 2) / (np.cbrt(x[-1] ** 3 + y[-1] ** 3) ** 2 + eps)
        dy = (y[-1] ** 2) / (np.cbrt(x[-1] ** 3 + y[-1] ** 3) ** 2 + eps)
        dx *= x[-1]/abs(x[-1])
        dy *= y[-1]/abs(y[-1])
        x.append(x[-1] - dx * step_size)
        y.append(y[-1] - dy * step_size)
        if np.sqrt(x[-1]**2 + y[-1]**2) < 1:
            break
        
    plt.plot(x, y, '.', markersize=5)

plt.xlabel("L3 regularization", fontsize=14)

# LInf
plt.subplot(224)
plt.axhline(y=0, color="black", linestyle="-")
plt.axvline(x=0, color="black", linestyle="-")
plt.xlim(-100, 100)
plt.ylim(-100, 100)

for i in range(len(init_x)):
    x, y = [init_x[i]], [init_y[i]]
    for _ in range(100):
        dx = x[-1]/abs(x[-1]) if abs(x[-1]) >= abs(y[-1]) else 0
        dy = 0 if abs(x[-1]) >= abs(y[-1]) else y[-1]/abs(y[-1])
        x.append(x[-1] - dx * step_size)
        y.append(y[-1] - dy * step_size)
        if np.sqrt(x[-1]**2 + y[-1]**2) < 1:
            break
    
    plt.plot(x, y, '.', markersize=5)

plt.xlabel("L-inf regularization", fontsize=14)
plt.savefig("lp_regularization.png", dpi=300)
plt.show()
```
