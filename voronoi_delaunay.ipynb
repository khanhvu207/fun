```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d, Delaunay, delaunay_plot_2d
```

```python
points = [
    [0, 0],
    [2, 0],
    [0, 1],
    [-1, 2]
]
```

```python
vor = Voronoi(points)
tri = Delaunay(points)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 13))
voronoi_plot_2d(vor, ax=ax)
delaunay_plot_2d(tri, ax=ax)

for p in points:
    x, y = p
    ax.plot(x, y, 'bo')

r = 2
for p in points:
    c = plt.Circle(tuple(p), r, color='g', fill=False)
    plt.gca().add_patch(c)

ax.set_xlim((-5,8))
ax.set_ylim((-5,6))
ax.grid()
plt.savefig("nemean_lion_00.pdf")
plt.show()
```
